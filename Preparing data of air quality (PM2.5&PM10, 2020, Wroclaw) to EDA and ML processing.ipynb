# Preparing data of air quality (PM2.5&PM10, 2020, Wroclaw) to EDA and ML processing

```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib as plt
import os
import datetime

%matplotlib inline
```

```python
#import data

path_list = []
for root, dirs, files in os.walk(".\Documents\Projekt_Alert_Smogowy\GIOS\Wyniki pomiarów z 2020 roku"):
    for name in files:
      path_list.append(os.path.join(root, name))

df_1 = pd.read_csv(path_list[0], index_col=False, engine='python', sep=';')
df_2 = pd.read_csv(path_list[1], index_col=False, engine='python', sep=';')
df_3 = pd.read_csv(path_list[2], index_col=False, engine='python', sep=';')
df_4 = pd.read_csv(path_list[3], index_col=False, engine='python', sep=';')
df_5 = pd.read_csv(path_list[4], index_col=False, engine='python', sep=';')
df_6 = pd.read_csv(path_list[5], index_col=False, engine='python', sep=';')
df_7 = pd.read_csv(path_list[6], index_col=False, engine='python', sep=';')
df_8 = pd.read_csv(path_list[7], index_col=False, engine='python', sep=';')
df_9 = pd.read_csv(path_list[8], index_col=False, engine='python', sep=';')
df_10 = pd.read_csv(path_list[9], index_col=False, engine='python', sep=';')
df_11 = pd.read_csv(path_list[10], index_col=False, engine='python', sep=';')
df_12 = pd.read_csv(path_list[11], index_col=False, engine='python', sep=';')
```

```python
df_1.drop(0, inplace = True)
df_2.drop(0, inplace = True)
df_3.drop(0, inplace = True)
df_4.drop(0, inplace = True)
df_5.drop(0, inplace = True)
df_6.drop(0, inplace = True)
df_7.drop(0, inplace = True)
df_8.drop(0, inplace = True)
df_9.drop(0, inplace = True)
df_10.drop(0, inplace = True)
df_11.drop(0, inplace = True)
df_12.drop(0, inplace = True)
```

```python
df_4.info()
```

```python
#concat data frames

df_list = [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12]

df_total = pd.concat(df_list)
```

```python
df_total.info()
```

```python
df_total.head(10)
```

```python
#sortin values by column Zanieczszczenie -> order dataset
df_total.sort_values(by='Zanieczyszczenie', inplace = True)
```

```python
#cleaning data
for x in range(1,6):
    df_total[df_total.columns[x]] = df_total[df_total.columns[x]].str.strip()
    
df_total.head()
```

```python
df_total[df_total.columns[1]].dtype
```

```python
df_total[df_total.columns[1]] = pd.to_numeric(df_total[df_total.columns[1]])
df_total[df_total.columns[2]] = pd.to_numeric(df_total[df_total.columns[2]])
df_total[df_total.columns[3]] = pd.to_numeric(df_total[df_total.columns[3]])
df_total[df_total.columns[4]] = pd.to_numeric(df_total[df_total.columns[4]])
df_total[df_total.columns[5]] = pd.to_numeric(df_total[df_total.columns[5]])
df_total.info()
```

```python
df_total.head()
```

```python
#change column with datatime

df_total.rename(columns={'Zanieczyszczenie': 'Date'}, inplace=True)
df_total.head()
```

```python
df_total[df_total.columns[0]] = pd.to_datetime(df_total[df_total.columns[0]])
df_total['Year'] = pd.DatetimeIndex(df_total['Date']).year
df_total['Month'] = pd.DatetimeIndex(df_total['Date']).month
df_total['Day'] = pd.DatetimeIndex(df_total['Date']).day
df_total['Hour'] = pd.DatetimeIndex(df_total['Date']).hour
df_total.head()
```

```python
df_total.info()
```

```python
#EDA

df_total.plot(x='Date', y=['Wrocław - Wiśniowa - pył zawieszony PM2.5', 
                           'Wrocław - Korzeniowskiego - pył zawieszony PM2.5'], kind='line', figsize=(20,10))
```
